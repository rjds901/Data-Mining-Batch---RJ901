```python
class DumData1:
    def __init__(self, unstackdata, stackdata, loop_no, loop_name):
        self.stackdata = stackdata
        self.loop_no = loop_no+1
        self.loop_name = loop_name
        self.unstackdata = unstackdata
        #self.tempdata = tempdata
        self.dum_cols = [f'{x}_{y}' for y in range(1, self.loop_no) for x in self.stackdata.columns[3:]]
        self.dum_cols = self.stackdata.columns[0:2].tolist() + self.dum_cols
        self.dum = pd.DataFrame(columns=self.dum_cols)
        
    def create_unstackdum1(self):
        nrows1 = self.unstackdata.shape[0]
        nrows2 = self.stackdata.shape[0]
        if nrows2>0:
            for i in range(nrows1):
                t2 = self.unstackdata.loc[i, 'respid']
                for j in range(nrows2):
                    if self.stackdata.loc[j, 'respid'] == t2:
                        t3 = self.stackdata.loc[j, self.loop_name]
                        for x in self.stackdata.columns[3:]:
                            self.dum.loc[i, f'{x}_{t3}'] = self.stackdata.loc[j, f'{x}']
        else:
            for i in range(nrows1):
                for j in range(nrows2):
                    t3 = self.stackdata.loc[j, self.loop_name]
                    for x in self.stackdata.columns[3:]:
                        self.dum.loc[i, f'{x}_{t3}'] = ''
            
        self.dum['responseid'] = self.unstackdata['responseid']
        self.dum['respid'] = self.unstackdata['respid']
        #self.dum = self.dum.loc[:, ~self.dum.columns.duplicated()]
        #self.dum.reset_index(drop=True, inplace=True)
        #print(self.dum.shape)
        if self.dum.shape[0]>0:
            self.unstackdata = pd.merge(self.unstackdata,self.dum,on =['responseid','respid'],how ='left')
        else:
            self.unstackdata = pd.concat(self.unstackdata,self.dum.iloc[2:])
        #print(self.unstackdata.shape)
    
        return self.unstackdata

    def create_unstackdum2(self):
            # Get matching rows from stackdata
        matching_rows = self.stackdata[self.stackdata['respid'].isin(self.unstackdata['respid'])]

        # Loop through unstackdata rows
        for i, row in self.unstackdata.iterrows():
            t2 = row['respid']
            
            # Get matching rows from stackdata for current respid
            t2_matching_rows = matching_rows[matching_rows['respid'] == t2]
            
            # Loop through columns of matching rows
            for col in t2_matching_rows.columns[3:]:
                # Apply function to get new column name
                col_name = f"{col}_{t2_matching_rows[self.loop_name].iloc[0]}"
                
                # Assign values to entire column at once using .loc
                self.dum.loc[i, col_name] = t2_matching_rows[col].values
            
        self.dum['responseid'] = self.unstackdata['responseid']
        self.dum['respid'] = self.unstackdata['respid']

        # Merge dum with unstackdata using .merge
        self.unstackdata = pd.merge(self.unstackdata, self.dum, on=['responseid', 'respid'], how='left')
        #print(self.unstackdata.shape)
        return self.unstackdata

```

```python
# Drop entire blank columns from the dataframe
df1.dropna(axis=1, how='all', inplace=True)
df1.shape
```

```python
# Split the dataframe into smaller dataframes with 5000 columns each
num_cols = len(df1.columns)
start_col = 0
dfs = []
while start_col < num_cols:
    end_col = min(start_col + 5000, num_cols)
    dfs.append(df1.iloc[:, start_col:end_col])
    start_col = end_col
```

```python
# Add the 'respid' column to all dataframes except the first one
for i in range(1, len(dfs)):
    dfs[i].insert(0, 'respid', df1['respid'])
```

```python
# Concatenate the smaller dataframes into a list of dataframes
small_dfs = []
for i in range(0, len(dfs)):
    small_dfs.append(dfs[i])
```

```python
# Display the first 5 rows of the first small dataframe
#print(small_dfs[0].head())
for i in range(len(small_dfs)):
    print('Sliced DataFram '+str(i+1)+" Shape : ", small_dfs[i].shape)
```

```python
# Write each sliced dataframe to a different worksheet in an Excel file
with pd.ExcelWriter('sliced_dataframes.xlsx') as writer:
    for i, dataframe in enumerate(dfs):
        dataframe.to_excel(writer, sheet_name=f'Data_{i+1}', index=False)

print("Done writing to Excel file")
```
